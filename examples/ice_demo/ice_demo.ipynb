```python
# Cell 1: imports
from qeplotter.qep import plot_from_file
```

```python
# Cell 2) Band structure for ICE (shifted so E_F = 0 eV)
plot_from_file(
    plot_type   = 'band',
    band_file   = 'ice.bands.dat.gnu',
    kpath_file  = 'ice.kpath',
    fermi_level = 1.5341,      # Fermi energy in eV
    shift_fermi = True,        # shift E so that E_F â†’ 0
    y_range     = (-10,10),
    dpi         = 300,
    savefig     = 'saved/ice_band.png'
)
```

```python
# Cell 3) Total DOS for ICE (shifted)
plot_from_file(
    plot_type   = 'dos',
    dos_file    = 'ice.dos',
    fermi_level = 1.5341,
    shift_fermi = True,
    y_range     = None,
    dpi         = 300,
    savefig     = 'saved/ice_dos.png'
)
```

```python
# Cell 4) Projected DOS (PDOS) by atom for ICE (shifted)
plot_from_file(
    plot_type   = 'pdos',
    pdos_dir    = 'ice_pdos',
    pdos_mode   = 'atomic',
    fermi_level = 1.5341,
    shift_fermi = True,
    y_range     = None,
    dpi         = 300,
    savefig     = 'saved/ice_pdos_atomic.png')
```
